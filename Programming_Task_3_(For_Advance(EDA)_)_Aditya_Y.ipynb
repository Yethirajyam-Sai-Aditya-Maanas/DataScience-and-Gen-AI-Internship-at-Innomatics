```python
#1️. Maximum Product of Two Elements in an Array
class Solution:
    def maxProduct(self, nums):
        max1 = 0
        max2 = 0

        for num in nums:
            if num > max1:
                max2 = max1
                max1 = num
            elif num > max2:
                max2 = num

        return (max1 - 1) * (max2 - 1)
```

```python
#2️. Count Number of Teams
class Solution:
    def numTeams(self, rating):
        n = len(rating)
        count = 0

        for j in range(n):
            left_smaller = 0
            left_larger = 0
            right_smaller = 0
            right_larger = 0

            for i in range(j):
                if rating[i] < rating[j]:
                    left_smaller += 1
                elif rating[i] > rating[j]:
                    left_larger += 1

            for k in range(j + 1, n):
                if rating[k] > rating[j]:
                    right_larger += 1
                elif rating[k] < rating[j]:
                    right_smaller += 1

            count += (left_smaller * right_larger)
            count += (left_larger * right_smaller)

        return count
```

```python
#3️. Number of Students Doing Homework at a Given Time
Problem:

Count students where:
startTime[i] <= queryTime <= endTime[i]

Time Complexity:

O(n)

class Solution:
    def busyStudent(self, startTime, endTime, queryTime):
        count = 0

        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                count += 1

        return count
```

```python
#4️. Number of Steps to Reduce a Number to Zero
class Solution:
    def numberOfSteps(self, num):
        steps = 0

        while num > 0:
            if num % 2 == 0:
                num //= 2
            else:
                num -= 1
            steps += 1

        return steps
```

```python
#5️. Counting Bits
class Solution:
    def countBits(self, n):
        result = [0] * (n + 1)

        for i in range(1, n + 1):
            result[i] = result[i // 2] + (i % 2)

        return result
```
